## Multiple Linear Regression

```python
df = pd.read_csv('data.csv', 
                 index_col='flight_temporal_order')
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt



fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =df['orings_at_thermal_distress']
y =df['temp_in_F']
z =df['leak_check_pressure_psi']



ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('orings_at_thermal_distress')
ax.set_ylabel('temp_in_F')
ax.set_zlabel('leak_check_pressure_psi')

plt.show()
```

##### Multiple Regression Analysis Interpretation

<br>The regression model accounts for 29.0% of the variance which is a bit low here.<br />
<br>Significance of F is very small here 0.0126 indicating a greater probability<br />
that the regression output is not by chance.<br />
<br>For every unit increase in temperature the orings at thermal distress decreases by 0.0487.<br />
There is a very low probability (0.9%) that the coeffcient is obtained by chance.<br />
<br>For every unit increase in pression the o-rings at thermal distress increase by 0.0029<br />
There is a 10.5% probability that the coefficient is obtained by chance.<br />

```python
est = smf.ols('df["orings_at_thermal_distress"] ~ df["temp_in_F"] + df["leak_check_pressure_psi"]',
              data=df).fit()
est.summary()
```

Predicted values of o-rings

```python
print('Predicted values: ', est.predict())
```

##### Polynomial regression

<br>1st order polynomial regression is nothing but a linear regression <br />
with an R square of 0.26 and shows about 2.3 o-rings at thermal distress at 31F<br />
<br>2nd order polynomial regression seems to be a better fit with an R square of 0.37 <br />
and shows about 6.5 o-rings at thermal distress as 31F<br />
<br>3rd order polynomial regression seems to be an overfit here with an R suare of 0.43 <br />
and a very high amount of o-rings at thermal distress at 31F<br />

```python
##plot o-rings at thermal distress using 1st, 2nd and 3rd order polynomials at 
##different temperatures

plt.figure(figsize=(6 * 1.618, 6))
plt.scatter(df.temp_in_F, df.orings_at_thermal_distress, s=30, alpha=1)
plt.xlabel('temp_in_F')
plt.ylabel('orings_at_thermal_distress')

# points linearlyd space on temperature
x = pd.DataFrame({'temp_in_F': np.linspace(30, df.temp_in_F.max(), 30)})



# 1-st order polynomial
poly_1 = smf.ols(formula='orings_at_thermal_distress ~ 1 + temp_in_F', data=df).fit()
plt.plot(x.temp_in_F, poly_1.predict(x), 'b-', label='Poly n=1 $R^2$=%.2f' % poly_1.rsquared, 
         alpha=0.9)

# 2-nd order polynomial
poly_2 = smf.ols(formula='orings_at_thermal_distress ~ 1 +  temp_in_F + I( temp_in_F ** 2.0)', 
                 data=df).fit()
plt.plot(x.temp_in_F, poly_2.predict(x), 'g-', label='Poly n=2 $R^2$=%.2f' % poly_2.rsquared, 
         alpha=0.9)



# 3-rd order polynomial
poly_3 = smf.ols(formula='orings_at_thermal_distress  ~ 1 +  temp_in_F + I( temp_in_F ** 2.0) + I( temp_in_F ** 3.0)', 
                 data=df).fit()
plt.plot(x.temp_in_F, poly_3.predict(x), 'r-', alpha=0.9,
         label='Poly n=3 $R^2$=%.2f' % poly_3.rsquared)


plt.legend()

##predicting o-rings at thermal distress using 2nd order polynomial
y_hat_int = poly_2.predict(x)

y_hat_int
```

## Multiple Linear Regression at different pressure values

<br>Here we carry out a multiple linear regression to predict o-rings 
showing thermal distress
at 31F and different pressures 0, 50, 100 and 200 PSI

Our plot predicts that at
<br>--- Temperature of 31F and pressure of 0psi there will be 1.75 o-rings at thermal distress
<br>--- Temperature of 31F and pressure of 50psi there will be 2.05 o-rings at thermal distress
<br>--- Temperature of 31F and pressure of 100psi there will be 2.2 o-rings at thermal distress
<br>--- Temperature of 31F and pressure of 2000psi there will be 2.45 o-rings at thermal distress

<br>This means that temperature does play a significant role in o-ring erosion.<br />
The lower the temperature higher the number of o-ring erosion.

<br>Temperature higher than 65F predict nearly 0 o-ring erosion but we have a few
cases showing high erosion at these temperatures. <br />
This means there are other factors at play here<br />


```python
plt.figure(figsize=(6 * 2, 10))
plt.scatter(df.temp_in_F, df.orings_at_thermal_distress, s=50, alpha=1)
plt.xlabel('temp_in_F')
plt.ylabel('orings_at_thermal_distress')

# points linearlyd space on temperature
x = pd.DataFrame({'temp_in_F': np.linspace(30, df.temp_in_F.max(), 100)})



est = smf.ols(formula='orings_at_thermal_distress ~ temp_in_F + leak_check_pressure_psi', data=df).fit()

plt.plot(x, est.params[0] + est.params[1] * x + est.params[2] * 0, 'r')
plt.plot(x, est.params[0] + est.params[1] * x + est.params[2] * 50, 'b')
plt.plot(x, est.params[0] + est.params[1] * x + est.params[2] * 100, 'g')
plt.plot(x, est.params[0] + est.params[1] * x + est.params[2] * 200, 'p')

est.params[0]
```
